```python
from linear_regression import LinearRegressionScratch
from sample_data_linear_reg import create_sample_data
from scaler import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Data 
X, y = create_sample_data()

df = pd.DataFrame({
    "X": X.flatten(),
    "y": y
})

df.head()
```

```python
# Linear Regression Implementation

# Load data
X, y = create_sample_data()

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train model
model = LinearRegressionScratch(learning_rate=0.1, n_iter=1000, verbose=True)
model.fit(X_scaled, y)

# Predictions
y_pred = model.predict(X_scaled)

# Evaluation
r2 = model.r2_score(y, y_pred)
print("R2 Score:", r2)
print("Predictions:", y_pred)

# Visualization
idx = np.argsort(X[:, 0])
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Actual data")
plt.plot(X[idx], y_pred[idx], color="red", label="Regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.show()

# Loss curve
plt.plot(model.losses)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss vs Iterations")
plt.show()
```
